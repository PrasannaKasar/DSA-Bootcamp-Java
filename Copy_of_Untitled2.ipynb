```python
from sklearn.preprocessing import OneHotEncoder
```

```python
onehotencoder=OneHotEncoder()
```

```python
career_encode=onehotencoder.fit_transform(x.iloc[:, 3:4])
```

```python
career_encoded_df = pd.DataFrame.sparse.from_spmatrix(career_encode, columns=onehotencoder.get_feature_names_out(['career']))
```

```python
x = pd.concat([x, career_encoded_df], axis=1)
```

```python
x.drop(['career'], axis=1, inplace=True)
```

```python
x
```

```python
imputer = SimpleImputer(missing_values=np.nan, strategy="mean")
imputer=imputer.fit(x.iloc[:, 8:9])
x.iloc[:,8:9]=imputer.transform(x.iloc[:,8:9])
```

```python
x.isnull().sum().sum()
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

```python
np.where(np.isnan(x_train))
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc_x = StandardScaler()
```

```python
x_train = sc_x.fit_transform(x_train)
```

```python
x_test = sc_x.transform(x_test)
```

```python
cls = svm.SVC(kernel='linear')
cls.fit(x_train, y_train)
pred = cls.predict(x_test)
accuracy = metrics.accuracy_score(y_test, y_pred=pred)
print('Accuracy of the model is =', accuracy*100, '%')
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm

X = x[['fun', 'attr']]
Y = y

clf = svm.SVC(kernel='rbf')
clf.fit(X, Y)

```

```python
x_min, x_max = X['attr'].min() - 1, X['attr'].max() + 1
y_min, y_max = X['fun'].min() - 1, X['fun'].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.02),
                     np.arange(y_min, y_max, 0.02))

Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

```

```python
plt.figure(figsize=(5, 5))
plt.contourf(xx, yy, Z, alpha=0.8, cmap=plt.cm.RdYlBu)
plt.scatter(X['fun'], X['attr'], c=y, cmap=plt.cm.RdYlBu, edgecolor='black')
plt.xlabel('Attractiveness')
plt.ylabel('Fun')
plt.title('SVM Decision Boundary: Matched vs Unmatched')

plt.scatter([], [], c='red', label='Unmatched')
plt.scatter([], [], c='blue', label='Matched')
plt.legend()

plt.show()
```
